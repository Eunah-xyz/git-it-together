```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'datas/DOB_Complaints_Received_20231117.csv'
complaints_df = pd.read_csv(path, engine = 'python') # df -> just helps me remember that it is a data frame
```

```python
complaints_df.head()
```

```python
complaints_df.Status.value_counts()
```

```python
active_complaints_df = complaints_df[[complaint for complaint in complaints_df.Status == 'ACTIVE']]
# look inside the complaints_df and look for activated statuse
```

```python
display(active_complaints_df)
```

```python
active_complaints_df['Complaint Category']
```

```python
illegal_social_club_df = complaints_df[[x for x in complaints_df['Complaint Category'] == '51']]
display(illegal_social_club_df)
```

```python
house_streets = illegal_social_club_df['House Street'].tolist()
# tolist() -> return python list instead of pandas objects
```

```python
house_streets_edit = [x.strip() for x in house_streets]
house_streets_edit
```

```python
house_numbers = illegal_social_club_df['House Number'].tolist()
house_numbers
```

```python
# zip -> place 2 list together
#[x,y for x,y in zip(dt_1, dt_2)]
bad_places = [num + ', ' + street for num, street in zip(house_numbers, house_streets_edit)]
bad_places
```

```python
# let's find the top 5 worst places are

from collections import Counter
worst_places = Counter(bad_places).most_common(5)
print(worst_places)
```
