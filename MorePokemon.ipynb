```python
# min() and max() function

weakest_pokemon = pokemon.hp.min()
strongest_pokemon = pokemon.hp.max()

print(weakest_pokemon)
print(strongest_pokemon)
```

```python
wimpy_pokemon = pokemon.loc[pokemon.hp == 1, 'name'].item() # .item() will give me just the value inside the cell
print(wimpy_pokemon)
```

```python
tough_pokemon = pokemon.loc[pokemon.hp == 255, 'name'].item()
print(tough_pokemon)
```

```python
# getting a subsection of the data frames

pokemon_army_idx = pokemon[[p for p in pokemon.hp > 150]] #don't use if. just use the name of the dataset
##[[]] -> data frames (list inside the list)
print(pokemon_army_idx)
```

```python
# let's make a graph!!
import matplotlib.pyplot as plt
```

```python
plt.hist(pokemon.hp, bins = 25) 
#hist -> histogram / bins -> how many graphs we get
plt.xlabel('HP')
plt.ylabel('Number of Pokemon')
plt.title('Distribution of Pokemon according to the HP')
```

```python
#grouping by types (grouped by each types... like water )
pokemon_grouping = pokemon.groupby('type1')
```

```python
water_pokemon = pokemon_grouping.get_group('water')
fire_pokemon = pokemon_grouping.get_group('fire')
```

```python
type(water_pokemon)
```

```python
plt.hist(water_pokemon.hp, bins = 20, label = 'water')
plt.hist(fire_pokemon.hp, bins = 20, label = 'fire')

plt.xlabel('HP')
plt.ylabel('Number of Pokemon')
plt.legend() # the types on top right corner
```
